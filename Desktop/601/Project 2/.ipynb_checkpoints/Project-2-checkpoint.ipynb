```python
import csv
import urllib.request
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.pyplot import pie, axis, show
%matplotlib inline
```

```python
#Reading the csv file from the URL and changing it into dataframe
pe_data=pd.read_csv('https://data.cityofnewyork.us/api/views/7uuj-b95m/rows.csv?accessType=DOWNLOAD')
```

```python
pe_data.shape
```

```python
pe_data #for reference
```

<center>Checking for data consistency</center>

```python
#checking the presence of null values in each column
pe_data.isnull().sum()
```

We can see that this data has no null values. Now checking if there is any column with all values as '0'.

```python
for column in pe_data:
    if((pe_data[column] == 0).all()):
        print (pe_data[column])
    else:
        print('none')
```

The data has no column with complete 0 values, which says the data is pretty clean. Using the 'describe()' function we can see the mathematical properties of data. 'mean' and 'std' can be used to observe the consistency of the data.

```python
pe_data.describe()
```

The data seems to be consistent as the standard deviation is less concerning the title of the column. Lesser standard deviation indicates that the values are closer to the mean value, which shows consistency in the data.

<font size=5>Analyzing and Visualizing the data!

Observation 1:Finding the number of Physical Education teachers in each Borough.

```python
pe_data.groupby('Borough')['Total Part Time and Full Time PE Teachers*'].sum().sort_values(ascending=False)
```

We can see that Brooklyn has the highers number of PE teachers enrolled.(Hypothesis 1 proven wrong)

Observation 2:Finding the number of students enrolled for PE.

```python
pe_data.groupby('School Level ')['Student Enrollment'].sum().sort_values(ascending=False)
```

We can see that Elementary school has the highest number of students enrolled.(Hypothesis 2 proven wrong)

Observation 3:Finding the maximum techer to student ratio.

```python
pe_data[pe_data['Ratio of Full Time Licensed PE Teachers to Students*']==
        pe_data['Ratio of Full Time Licensed PE Teachers to Students*'].max()]
```

High school and secondary school have the highest teacher to student ratio. (Hypothesis 3 proven wrong)

observation 4: Finding number of schools having no PE teacher

```python
no_PE=(pe_data['Total Part Time and Full Time PE Teachers*']==0).sum()
percentage_noPE=no_PE/pe_data.shape[0]*100
print("Number of schools having no PE teachers: "+str(no_PE))
print("Percentage of schools having no PE teacher: "+str(percentage_noPE)+"%")
```

Observation 5:Visualizing the PE teacher count w.r.t school level.

```python
pe_data.groupby(['School Level ']).count()['Total Part Time and Full Time PE Teachers*'].plot()
plt.xticks(rotation=90)
plt.ylabel('number of PE teachers')
```

We can see that elementary school has the highest enrolled PE teachers.

Observation 6:Visualizing the percentage of student enrollment in each Borough.

```python
sums=pe_data.groupby('Borough').sum()['Student Enrollment']
axis('equal')
plt.title('Students enrolled in each Borough')
plt.pie(sums, labels=sums.index,autopct='%1.1f%%',radius=1.3)
plt.show()
```

We can see that Brooklyn and Queens have almost the same percentage of students enrolled but Queens leading the chart with 29.2%
